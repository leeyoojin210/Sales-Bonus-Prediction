```python
# Outlier treatment of required
import pandas as pd
import numpy as np

# Load the data from the Excel file
df = pd.read_excel('Sales.xlsx', "Sales")
# Select the required columns for outlier treatment
columns_of_interest = ['AgentBonus', 'Age', 'CustTenure', 'MonthlyIncome', 'Complaint', 'CustCareScore']

# Define the outlier treatment function
def treat_outliers(column):
    q1 = np.percentile(column, 25)
    q3 = np.percentile(column, 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    column = np.where((column < lower_bound) | (column > upper_bound), np.median(column), column)
    return column

# Apply outlier treatment to the selected columns
for col in columns_of_interest:
    df[col] = treat_outliers(df[col])

# Verify the changes
print(df[columns_of_interest].describe())
```

```python
data=pd.read_excel('Sales.xlsx',sheet_name='Sales')
data.Channel.unique()
```

```python
# data=pd.read_csv('CE802_P3_Data.csv')

def channel(s):
    if s=='Agent':
        return 1
    elif s=='Third Party Partner':
        return 2
    elif s=='Online':
        return 3

def occupation(s):
    if s=='Salaried':
        return 1
    elif s=='Free Lancer':
        return 2
    elif s=='Small Business':
        return 3
    elif s=='Laarge Business':
        return 4
    elif s=='Large Business':
        return 5

def education(s):
    if s=='Graduate':
        return 1
    elif s=='Post Graduate':
        return 2
    elif s=='UG':
        return 3
    elif s=='Under Graduate':
        return 4
    elif s=='Engineer':
        return 5
    elif s=='Diploma':
        return 6
    elif s=='MBA':
        return 7


def gender(s):
    if s=='Female' or s=='Fe male':
        return 1
    elif s=='Male':
        return 2

def designation(s):
    if s=='Manager':
        return 1
    elif s=='Exe':
        return 2
    elif s=='Executive':
        return 3
    elif s=='VP':
        return 4
    elif s=='AVP':
        return 5
    elif s=='Senior Manager':
        return 6


def maritalStatus(s):
    if s=='Single':
        return 1
    elif s=='Divorced':
        return 2
    elif s=='Unmarried':
        return 3
    elif s=='Married':
        return 4

def zone(s):
    if s=='North':
        return 1
    elif s=='West':
        return 2
    elif s=='East':
        return 3
    elif s=='South':
        return 4

def paymentMethod(s):
    if s=='Half Yearly':
        return 1
    elif s=='Yearly':
        return 2
    elif s=='Quarterly':
        return 3
    elif s=='Monthly':
        return 4

data['Channel']=data['Channel'].apply(channel)
data['Occupation']=data['Occupation'].apply(occupation)
data['EducationField']=data['EducationField'].apply(education)
data['Gender']=data['Gender'].apply(gender)

data['Designation']=data['Designation'].apply(designation)
data['MaritalStatus']=data['MaritalStatus'].apply(maritalStatus)
data['Zone']=data['Zone'].apply(zone)

data['PaymentMethod']=data['PaymentMethod'].apply(paymentMethod)
```

```python
Input_data=data[['CustID',  'Age', 'CustTenure', 'Channel', 'Occupation',
       'EducationField', 'Gender', 'ExistingProdType', 'Designation',
       'NumberOfPolicy', 'MaritalStatus', 'MonthlyIncome', 'Complaint',
       'ExistingPolicyTenure', 'SumAssured', 'Zone', 'PaymentMethod',
       'LastMonthCalls', 'CustCareScore']]
Target=data[['AgentBonus']]
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Input_data, Target, test_size=0.33, random_state=42)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
clf1 = LinearRegression()
clf1.fit(X_train,y_train)
pred=clf1.predict(X_test)
linearRegressionScore=clf1.score(X_test,y_test)
print("LinearRegression Testing accuracy: ",linearRegressionScore)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, pred)
plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (pred)')
plt.title('Linear Regression: Actual vs. Predicted')
plt.show()
```
