## Scatter plot

```python
# use scatter_matrix of pandas.plotting
# https://www.geeksforgeeks.org/pair-plots-using-scatter-matrix-in-pandas/
from pandas.plotting import scatter_matrix
scatter_matrix(data[['CustID', 'AgentBonus', 'Age', 'CustTenure', 'ExistingProdType',
       'NumberOfPolicy', 'MonthlyIncome', 'Complaint',
       'ExistingPolicyTenure', 'CustCareScore']], figsize=(16, 12))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Calculate percentage counts of each classification type
classification_counts = data['Channel'].value_counts(normalize=True) * 100

# Plot the bar graph with percentage counts
plt.figure(figsize=(12, 6))
classification_counts.plot(kind='bar')
plt.title('Percentage Count of Each Channel')
plt.xlabel('Channel')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
print(data.describe())  # it gives you the data of all ranges of boxplot
data.boxplot(figsize=(16, 8))
plt.title("Box Plot")
plt.xlabel("Column Names")
plt.ylabel("Outliers value")
plt.show()
```

```python
# From the above graph it can be concluded that there are some outliers in the SumAssured.
```

## Correlation

Correlation coefficient measures linear relationship between a pair of variables, range [-1, 1]. 1 indicates full linear relationship, 0 no linear relationship, -1 negatively linear relationship.

```python
# https://seaborn.pydata.org/generated/seaborn.heatmap.html
# to visualize the correlation between different columns in the dataset
import seaborn as sns
sns.set(rc={'figure.figsize':(12,10)})
corr_matrix = data.corr()
sns.heatmap(corr_matrix,annot=True)
plt.title("HeatMap Plot")
plt.xlabel("Column Names")
plt.ylabel("Column Names")
plt.show()
```
